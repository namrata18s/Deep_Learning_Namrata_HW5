```python
#load the CIFAR-10 dataset

import torchvision.transforms as T
from torchvision import datasets
from torch.utils.data import DataLoader

transform = T.Compose([
    T.ToTensor(),
])

dataset = datasets.CIFAR10(root='data', download=True, transform=transform)
loader = DataLoader(dataset, batch_size=64, shuffle=True)
```

```python
# import the DDPM model (from lucidrains) 

from denoising_diffusion_pytorch import Unet, GaussianDiffusion

model = Unet(
    dim=64,
    dim_mults=(1, 2, 4)
)

diffusion = GaussianDiffusion(
    model,
    image_size=32,
    timesteps=1000,  # DDPM default
    sampling_timesteps=200
).cuda()
```

```python
#train your model

import torch

optimizer = torch.optim.Adam(diffusion.parameters(), lr=1e-4)

for epoch in range(50):  # adjust based on GPU time
    for imgs, _ in loader:
        imgs = imgs.cuda()
        loss = diffusion(imgs)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    print(f"Epoch {epoch} | Loss: {loss.item():.4f}")
```

```python
import os
from torchvision.datasets import CIFAR10

# Load CIFAR-10 test set (its already downloaded if you used it before)
testset = CIFAR10(root="data", train=False, download=True)

# New folder to save PNG images
save_dir = "cifar10_test_images"
os.makedirs(save_dir, exist_ok=True)

print("Saving CIFAR-10 test images to:", save_dir)

for i, (img, label) in enumerate(testset):
    img.save(os.path.join(save_dir, f"{i:05d}.png"))

print("Done! Total images saved:", len(testset))
```
